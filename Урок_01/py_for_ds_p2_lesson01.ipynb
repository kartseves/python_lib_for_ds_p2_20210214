## Вопрос 1

### Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

#### Ошибки бывают двух видов: ложноположительные (FP) и ложноотрицательные (FN). И в случае если один из этих типов ошибок нам более страшен - мы постараемся его не допускать, даже в ущерб ошибкам другого типа.

#### Если мы пытаемся как можно реже делать ложноположительные предсказания, т.е. важно минимизировать FP, то мы максимизируем метрику precision.

Например, при фильтрации спама, когда фильтры спама или методы блокировки спама ошибочно классифицируют порядочное письмо как спам и, как следствие, препятствуют его прочтению.

 Или, например ложноположительное предсказание - это слишком дорого, когда мы ищем где добывать нефть.

#### Если один из типов ошибок нам более страшен и нам важно его не допускать, даже в ущерб ошибкам другого типа, т.е. важно минимизировать FN, то мы максимизируем метрику recall.

Например, при медицинском тестировании false negative будет давать ложно обнадеживающее сообщение пациентам и врачам о том, что болезнь отсутствует, когда в действительности она присутствует. Это приводит к неуместному или неадекватному лечению, или его отсутствию. Таким образом, в этом случае лучше false positive.

Или, например, нам важно не допустить на секретный объект нелегальных лиц.

## Вопрос 2

### Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

#### F-мера представляет собой гармоническое среднее между precision и recall. Она стремится к нулю, если любой из этих параметров стремится к нулю. Эту метрику мы хотим видеть высокой если нам важен баланс Precision и Recall.

Т.е. для F-меры важно чтобы обе метрики были как можно больше 0, для среднего арифметического же допускается "жертва" в пользу одной из двух.
